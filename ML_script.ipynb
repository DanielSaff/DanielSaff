```python
###determine hyperparmeters with GridSearchCV

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report


#load dataset
dataset_import = pd.read_csv(r'/Users/danielsaffertmuller/Documents/Gesundheits- und Rehabilitationstechnik /Masterarbeit/Cereneo/dataprocessing/matlab_functions/Features_prod.csv')
dataset = np.array(dataset_import)

#show shape of dataset
print(f'Shape of Dataset{dataset.shape}')



#show shape of features
x=dataset[:,:56]
print(f'Shape of Features{x.shape}')

#show shape of class
y=dataset[:,56]
print(f'Shape of class{y.shape}')
print(f'class{y}')


groups=dataset[:,57]
print(f'Shape of class{groups.shape}')
print(f'group{groups}')



# Split the dataset in two equal parts
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=0)

# Set the parameters by cross-validation
tuned_parameters = [
    {"max_features": [1,20],
     "max_depth": [1,10],
     "class_weight": ["balanced", "balanced_subsample"],
     "n_estimators": [50, 60, 70, 80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, 200]}
    
]


scores = ["precision", "recall"]

for score in scores:
    print("# Tuning hyper-parameters for %s" % score)
    print()

    clf = GridSearchCV(RandomForestClassifier(), tuned_parameters, scoring="%s_macro" % score)
    clf.fit(X_train, y_train)

    print("Best parameters set found on development set:")
    print()
    print(clf.best_params_)
    print()
    print("Grid scores on development set:")
    print()
    means = clf.cv_results_["mean_test_score"]
    stds = clf.cv_results_["std_test_score"]
    for mean, std, params in zip(means, stds, clf.cv_results_["params"]):
        print("%0.3f (+/-%0.03f) for %r" % (mean, std * 2, params))
    print()

    print("Detailed classification report:")
    print()
    print("The model is trained on the full development set.")
    print("The scores are computed on the full evaluation set.")
    print()
    y_true, y_pred = y_test, clf.predict(X_test)
    print(classification_report(y_true, y_pred))
    print()
```

```python
#k_fold cross validation

#load dataset
dataset_import = pd.read_csv(r'/Users/danielsaffertmuller/Documents/Gesundheits- und Rehabilitationstechnik /Masterarbeit/Cereneo/dataprocessing/matlab_functions/Features_prod_v2.1.csv')
dataset = np.array(dataset_import)

#show shape of dataset
print(f'Shape of Dataset{dataset.shape}')



#show shape of features
x=dataset[:,:42]
print(f'Shape of Features{x.shape}')

#show shape of class
y=dataset[:,42]
print(f'Shape of class{y.shape}')
print(f'class{y}')


groups=dataset[:,43]
print(f'Shape of class{groups.shape}')
print(f'group{groups}')




from sklearn.model_selection import GroupKFold

clf = RandomForestClassifier(class_weight= "balanced",max_features= 4, max_depth=9, n_estimators= 200)


gkf = GroupKFold(n_splits=8)
for train, test in gkf.split(x, y, groups=groups):
        x_train=x[train]
        y_train=y[train]
        print(x_train.shape)
        print(y_train.shape)
        
        clf.fit(x_train, y_train)
        
        y_pred=clf.predict(x[test])
        y_test=y[test]
        print(y_pred.shape)
        print(y_test.shape)
        
        score= clf.score(x[test], y[test])
        print(score)
        print(confusion_matrix(y_test, y_pred))
        print(classification_report(y_test, y_pred))
        
```
